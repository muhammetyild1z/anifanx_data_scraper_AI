```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests
import pyautogui
import pyperclip
import re


from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
url=''
driver = webdriver.Chrome()  
driver.get(url)
driver.maximize_window()  
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')
anime_name_text=soup.select_one('body > div.container > div > div.app-wrapper > div.app-container.flex-fill > div > div > div.detail-content > div.detail-text.flex-fill > div.caption > div.caption-content > a > h1').text.replace(":","").replace("\n","").replace(" ","_")
# Find the element containing the episode links
div_element1 = soup.find_all('a', id=re.compile("^season-\d+-tab$"))
if div_element1:  
    for anchor_element1 in div_element1:
        replaceElement = anchor_element1['id'].replace("-tab", "")      
        div_element = soup.find(id=replaceElement)        
        anchor_elements = div_element.find_all("a")
        
        # Open the file in append mode
        with open(f"{anime_name_text}url.txt", "a", encoding="utf-8") as file:
            # Iterate over each episode link
            for anchor_element in anchor_elements:
                href_value = anchor_element.get("href")
                if href_value:
                    file.write(href_value + "\n")  

Sibbnet_save_url=""
OkRu_save_url=""
odnoklassniki_save_url=""
VK_save_url=""

with open(f"{anime_name_text}url.txt", "r", encoding="utf-8") as file:
    url_list = file.readlines()
for url in url_list:
    url = url.strip()
    driver.get(url)
    # Oynatıcı seçimini metin olarak al
    playerCheck = soup.select_one('body > div.container > div > div.app-wrapper > div.app-container.flex-fill > div > div > div.detail-header.d-flex.align-items-center > div.nav-player-select.dropdown').text
    #sibnet linki 
    if "Sibnet" in playerCheck:
        driver.find_element(By.ID, "videoSource").click()
        
        # Tüm düğmeleri bulmak için bekleyin
        sibnet_button = WebDriverWait(driver, 1).until(
            EC.element_to_be_clickable((By.XPATH, "//span[text()='Sibnet']"))
        )
        sibnet_button.click()
        time.sleep(2)
        play_button = driver.find_element(By.XPATH, "//div[@class='play-btn']").click()
        time.sleep(5)
        pyautogui.click(x=1022, y=400)
        time.sleep(1)  
        pyautogui.click(x=1600, y=410 )
        time.sleep(1) 
        pyautogui.click(x=1259, y=700 )
        time.sleep(1) 
        pyautogui.click(x=1259, y=720 )       
        # Butona basıldığında panoya kopyalanan veriyi al
        sibnetUrl = pyperclip.paste()
        url_start_index = sibnetUrl.find('https')
        url_end_index = sibnetUrl.find('&', url_start_index)
        sibnet_save_url = sibnetUrl[url_start_index:url_end_index]
        pyperclip.copy('')
        
    if "Ok.ru" in playerCheck:
        driver.find_element(By.ID, "videoSource").click()
        
        # Tüm düğmeleri bulmak için bekleyin
        ok_ru_button = WebDriverWait(driver, 1).until(
            EC.element_to_be_clickable((By.XPATH, "//span[text()='Ok.ru']"))
        )
        ok_ru_button.click()
        time.sleep(2)  
        play_button = driver.find_element(By.XPATH, "//div[@class='play-btn']").click()
        time.sleep(3)
        pyautogui.click(x=1022, y=400)
        time.sleep(6)  
        pyautogui.click(x=1622, y=412 )
        pyautogui.click(x=1622, y=412 )
        time.sleep(1)
        pyautogui.click( x=1072, y=610 )
        time.sleep(1)
        pyautogui.click( x=1072, y=775 )
         # Butona basıldığında panoya kopyalanan veriyi al
        OkRuUrl = pyperclip.paste()
        url_start_index = OkRuUrl.find('//ok.ru')
        url_end_index = OkRuUrl.find('"', url_start_index)
        OkRu_save_url = OkRuUrl[url_start_index:url_end_index]
        pyperclip.copy('')  
        
    if "Odnoklassniki" in playerCheck:
        driver.find_element(By.ID, "videoSource").click()
        
         # Tüm düğmeleri bulmak için bekleyin
        odnoklassniki_button = WebDriverWait(driver, 1).until(
            EC.element_to_be_clickable((By.XPATH, "//span[text()='Odnoklassniki']"))
        )
        odnoklassniki_button.click()
        time.sleep(2)
        play_button = driver.find_element(By.XPATH, "//div[@class='play-btn']").click()
        time.sleep(4)
        pyautogui.click(x=1022, y=400)
        time.sleep(6)
        pyautogui.click(x=1622, y=412 )
        pyautogui.click(x=1622, y=412 )
        time.sleep(1)
        pyautogui.click( x=1072, y=610 )
        time.sleep(1)
        pyautogui.click( x=1072, y=775 )
         # Butona basıldığında panoya kopyalanan veriyi al
        odnoklassniki = pyperclip.paste()
        url_start_index = odnoklassniki.find('//odnoklassniki')
        url_end_index = odnoklassniki.find('"', url_start_index)
        odnoklassniki_save_url = odnoklassniki[url_start_index:url_end_index]
        pyperclip.copy('')   
        
    if "VK" in playerCheck:
        driver.find_element(By.ID, "videoSource").click()
        
         # Tüm düğmeleri bulmak için bekleyin
        VK_button = WebDriverWait(driver, 1).until(
            EC.element_to_be_clickable((By.XPATH, "//span[text()='VK']"))
        )
        VK_button.click()
        time.sleep(2)
        play_button = driver.find_element(By.XPATH, "//div[@class='play-btn']").click()
        time.sleep(8)
        pyautogui.rightClick(x=1022, y=400)
        time.sleep(1)
        pyautogui.click(x=1045, y=485)
        VK = pyperclip.paste()
        url_start_index = VK.find('https://vk.co')
        url_end_index = VK.find('"', url_start_index)
        VK_save_url = VK[url_start_index:url_end_index]
        
        
        episode_name=soup.select_one('body > div.container > div > div.app-wrapper > div.app-container.flex-fill > div > div > div.detail-content > div.detail-text.flex-fill > div.caption > div.caption-content > h2').text
        anime_name=soup.select_one('body > div.container > div > div.app-wrapper > div.app-container.flex-fill > div > div > div.detail-content > div.detail-text.flex-fill > div.caption > div.caption-content > a > h1').text.replace(":","")
        anime_seasons_episodes = soup.select_one('body > div.container > div > div.app-wrapper > div.app-container.flex-fill > div > div > div.detail-content > div.detail-text.flex-fill > div.caption > div.caption-content').text
        anime_seasons_episodes_array=anime_seasons_episodes.split()
        sezon=anime_seasons_episodes_array[0].replace(".","")
        episode=anime_seasons_episodes_array[2].replace(".","")
        anime_file_text_name=anime_name.replace(":","").replace("\n","").replace(" ","_")
        # Veriyi bir metin dosyasına yaz
        with open(f"{anime_file_text_name}details.txt", "a", encoding="utf-8") as file:
            file.write(anime_name+"\t"+episode_name+"\t"+sezon+"\t"+episode+"\t"+sibnet_save_url+"\t"+OkRu_save_url+"\t"+odnoklassniki_save_url+"\t"+VK_save_url+"\n")

    
        print("Veriler başarıyla kopyalandı.")
      
# Tarayıcıyı kapat
driver.quit()
```
